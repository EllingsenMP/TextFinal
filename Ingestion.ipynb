```python
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
import jinja2
```

```python
lyrics = pd.read_csv('gaga_tracks.csv').dropna()
lyrics.head()
```

```python
streams = pd.read_csv('gaga_streams.csv')
streams['track_name'] = streams['track_name'].str.lower()
lyrics['track_name'] = lyrics['track_name'].str.lower()
data = pd.merge(lyrics, streams, 'left', 'track_name', )
```

```python
def clean_lyrics(raw: str) -> str:
    lines = raw.split("\n")
    cleaned = []
    in_lyrics = False

    for line in lines:
        raw = raw.replace('\\n', '\n')
    
    raw = re.sub(r'[\u2005\u00a0\u2009\u202f\u2003\u2002]', ' ', raw)
    raw = raw.strip('"')
    raw = re.sub(r'\[.*?\]\n?', '', raw)
    raw = re.sub(r'[^\w\s]', '', raw)
    raw = re.sub(r'\n{3,}', '\n\n', raw)
    raw = re.sub(r'\n', ' ', raw)
    raw = raw.lower()
    
    return raw.strip()

data['lyrics'] = data['lyrics'].apply(clean_lyrics)
```

```python
albumList = [
    'MAYHEM',
    'Harlequin',
    'Chromatica',
    'A Star Is Born Soundtrack (Without Dialogue)',
    'Joanne (Deluxe)',
    'ARTPOP',
    'Born This Way',
    'The Fame',
    'The Fame Monster (Deluxe Edition)'
]

data = data[data['album'].isin(albumList)]
data['album'].unique()
```

```python
data = data[~data['track_name'].str.contains("alibi", case=False, na=False)]
data = data[~data['track_name'].str.contains("radio edit", case=False, na=False)]
data = data[~data['track_name'].str.contains("too far gone", case=False, na=False)]
data = data[~data['track_name'].str.contains("sour candy", case=False, na=False)]
data = data[~data['artists'].str.contains("bradley cooper", case=False, na=False)]

data = data.reset_index()

data = data[data['index'] != 240]
data.sort_values(by = 'streams', ascending = False).head(5)
```
